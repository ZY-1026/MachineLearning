# 基本概念(假设数据线性可分)


决策边界不唯一(不适定问题)

寻找好的决策边界，使得模型的泛化能力较强，使得决策边界距离两种分类点的距离要尽可能远



寻找距离决策边界最近的点，使得这些最近的点距离决策边界要尽可能的远



SVM尝试寻找一个最优的决策边界，距离两个类别的最近样本点最远，决策边界和L1和L2平行，在L1和L2之间将不存在其他点，L1和L2称为支撑向量，决策边界位于L1和L2之间

对于线性可分的数据，一般将SVM称为Hard margin svm,对于线性不可分的数据，将SVM称为Soft Margin SVM
